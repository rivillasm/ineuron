```python
# instantiate a Pipeline object that points to a directory on your file system:
import Augmentor
p = Augmentor.Pipeline("C:\\Users\\rivil\\DataCampINeuron\\General\\CNN\\MASK_RCNN\\Augmentor\\test\\")
```

```python
# Every function requires you to specify a probability, which is used to decide if an operation is 
# applied to an image as it is passed through the augmentation pipeline.

# probability=0.7 means 70% of the images are being applied this procedure

p.rotate(probability=0.7, max_left_rotation=10, max_right_rotation=10)
```

```python
p.rotate90(probability=0.5)
p.rotate270(probability=0.5)
p.flip_left_right(probability=0.75)
p.flip_top_bottom(probability=0.5)
p.crop_random(probability=1, percentage_area=0.5)
p.resize(probability=1.0, width=80, height=80)
```
