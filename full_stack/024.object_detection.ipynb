### Common components available in almost all object detection model/architecture 

* Region proposal - DL model which will locate the region of interest (ROJ)
* Feature extraction - & Network prediction
* non-maximum separation (NMS) - solves the issue of overlapping Bounding Boxes BBox , from several boxes it keeps the one with better score 



### Region Proposal   - ROI region the system believes have a high likehood of containing an object teremined by object score

### Feature extraction & network prediction - based on region proposal : get the box and do the prediciton of the class
* Bounding box prediction

### NMS - non maximum separation is a technique that makes sure you also deterct each object only once
* it finds out box with maximum prediction probability and eliminates other boxes
* discard all the bboxes with threshold less than given confidence threshold




* uses intersection over union (IOU) to discard Bbox which has low IOU score
