# face Detection and tracking from a frame



https://github.com/opencv/opencv/tree/master/data/haarcascades

* you need to convert the color image to grayscale because the HaarCascadeFace algorithm was trained on grayscale image

```python
import cv2
```

```python
alg = "haarcascade_frontalface_default.xml"  # face detection model
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+alg)    # load the model
cam = cv2.VideoCapture(0)                    # initializae the camera


while(True):
    _,img = cam.read()                                     # read the frame from the camera
    grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)         # converts color to gray
    face = haar_cascade.detectMultiScale(grayImg,1.3,4)    # get coordinates of the face

    for (x,y,w,h) in face:                                  # segregating x,y,h,w
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),2)   # color and thickness (0,255,0),2
   
    cv2.imshow("FaceDetection",img)
    
    # set the quit keyboard
    key = cv2.waitKey(1) 
    if key == 27:
        break
        
cam.release()
cv2.destroyAllWindows()
```
