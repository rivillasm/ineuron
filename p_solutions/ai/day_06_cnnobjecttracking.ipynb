# OBJECT TRACKING



* circle around the object



* image moments - to find the center of the object



* drawing circles


* workflow



* the job flow provide instructions
    * when an object is to close, the circle is large then the stop command is given by the program
    * when the object is at the left, the command to go right is given
    * when the object is at the right, the command to go left is given
    * the the circle of the object is small, the object is considered to be far and the commang to go forward is given

```python
import cv2
import imutils #resizing library
```

```python
redLower = (149,80,0)
redUpper = (166,255,255)
# redLower = (15,164,162)
# redUpper = (45,232,255)

cam = cv2.VideoCapture(0)                    # initializae the camera


while(True):
    (grabbed, frame) = cam.read()                          # read the frame from the camera

    frame = imutils.resize(frame, width=600)               # resize the frame
    blurred = cv2.GaussianBlur(frame,(11,11),0)            # smooth apply
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)         # convert to hsv from gbr
    
    
    mask = cv2.inRange(hsv, redLower, redUpper)            # mask the given color
    mask = cv2.erode(mask, None, iterations=2)             # 
    mask = cv2.dilate(mask, None, iterations=2)
    
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
    center = None
    
    
    if len(cnts) > 0:
        c = max(cnts, key=cv2.contourArea)
        ((x,y),radius) = cv2.minEnclosingCircle(c)      # gets x,y,radius of the circle
        M = cv2.moments(c)
        center = (int(M["m10"] / M["m00"]), int(M["m01"] / M["m00"]))
        
        if radius > 10:
            cv2.circle(frame, (int(x), int(y)), int(radius), (0,255,255), 2)
            cv2.circle(frame, center, 5, (0,0,255), -1)     # red color dot for left/right location

            if radius > 250:
                print("Stop")
            else:
                if(center[0]<150):
                    print("Left")
                elif(center[0]>450):
                    print("Right")
                elif(radius<250):
                    print("Front")
                    
    cv2.imshow("Frame",frame)
    
    # set the quit keyboard
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
        
cam.release()
cv2.destroyAllWindows()
```
