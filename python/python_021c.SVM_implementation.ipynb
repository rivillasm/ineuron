```python
columns=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
```

```python
#from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler   # works better with outliers
#scaler = StandardScaler()
scaler = RobustScaler()
df_sc = scaler.fit_transform(df.drop(labels=["quality"],axis=1))  # outputs an array without columns info
df_sc = pd.DataFrame(df_sc, columns=columns)
df_sc
```

```python
X=df_sc.values
print(X.shape)
X
```

```python
y=df["quality"]
print(y.shape)
y
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.33,random_state=42)
print(train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(train_x,train_y)
```

```python
pred_y = model.predict(test_x)
pred_y
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

accuracy_score(test_y,pred_y)
```

```python
print(classification_report(test_y, pred_y))
```

$Notice$ that recall and precision for some classes are always 0. It means that the classifier is always classifying everything into a single class i.e class 1! This means our model needs to have its parameters tuned.

```python
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear','poly','rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(train_x,train_y)
```

```python
# print best parameter after tuning
print(grid.best_params_)
 
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
model_bestparm=SVC(C= 1, gamma= 1, kernel= 'rbf')  # model with best parameters
model_bestparm.fit(train_x,train_y)
```

```python
pred_y = model_bestparm.predict(test_x)   # new predictions with the new best parameters
pred_y
```

```python
print(classification_report(test_y, pred_y))
```
