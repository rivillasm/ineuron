# Data Distribution


```python
import numpy as np
x=np.random.randn(10000)
x
```

```python
sns.distplot(x,kde=True, hist=True, rug=False, bins=30)
```

```python
tips
```

```python
x=tips.total_bill
y=tips.tip
sns.jointplot(x=x,y=y)
```

```python
sns.jointplot(x=x,y=y, kind='kde', shade=True)
```

```python
sns.kdeplot(x,y, shade=True, n_levels=60)
```

```python
import matplotlib.pyplot as plt
g=sns.jointplot(x,y,kind='kde', color='r')
g.plot_joint(plt.scatter, c='b', s=30, linewidth=1, marker='+')
g.ax_joint.collections[0].set_alpha(1)
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)    # creates the grid
g.map_diag(sns.kdeplot)   # kde plot on diagonal
g.map_offdiag(sns.kdeplot, n_levels=10)  # plot the diagonals
```
